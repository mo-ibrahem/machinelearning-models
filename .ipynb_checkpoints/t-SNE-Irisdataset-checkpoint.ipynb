```python
# Improting Modules
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# Loading Dataset
iris_df = datasets.load_iris()
```

```python
# Defining the Model
model = TSNE(learning_rate = 100)
```

```python
# fitting the Model
irisModel = model.fit_transform(iris_df.data)
```

```python
# plotting 2d t-SNE
x_axis = irisModel[:, 0]
y_axis = irisModel[:, 1]
```

```python
plt.scatter(x_axis, y_axis, c = iris_df.target)
plt.show()
```
